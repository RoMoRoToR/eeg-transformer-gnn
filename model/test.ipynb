```python
from torch_geometric.data import Data
from graphgps.network.gps_model import GPSModel

# Создаём тестовый граф
edge_index = torch.tensor([[0, 1, 1, 2], [1, 0, 2, 1]], dtype=torch.long)
x = torch.tensor([[1], [2], [3]], dtype=torch.float)

# Задаём графовые данные
graph = Data(x=x, edge_index=edge_index)

# Модель GPS (Graph Pointer Transformer)
model = GPSModel(num_layers=4, dim_hidden=64, num_heads=4)

# Прогоняем данные через модель
out = model(graph)
print(out.shape)
```
